```python
import numpy as np
import matplotlib.pyplot as plt
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
```

```python
# set width of bars
barWidth = 0.25
 
# set heights of bars

bars1 = [4713, 3871, 841]
bars2 = [7448, 200, 0,]
bars3 = [17000, 0 , 0]


 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='XDBX')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='SparkSQL')
plt.bar(r3, bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='Spark + Write')
 
# Add xticks on the middle of the group bars
plt.ylabel('Runtime', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Total', 'Execution', 'Registration'])
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
